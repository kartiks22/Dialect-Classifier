```python
from python_speech_features import logfbank
import scipy.io.wavfile as wav
import pandas as pd
import numpy as np
```

```python
for i in range(1,10):
    for j in range(1,68):
        (rate,sig) = wav.read("Read_Up/IDR"+str(i)+"/"+str(j)+".wav")
        fbank_feat = logfbank(sig,rate)
        temp = np.array(fbank_feat.mean(0))
        if i==1 and j==1:
            data = np.append(temp,i)
        else:
            temp = np.append(temp,i)
            data = np.vstack([temp, data])
```

```python
df = pd.DataFrame(data, dtype='float')
df.head()
```

```python
from sklearn.utils import shuffle
shuffled = shuffle(df)
shuffled.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(shuffled, test_size = 0.3)
```

```python
x_train = train.iloc[1:423, 1:26]
y_train = train.iloc[1:423, 26:27]
x_test = test.iloc[1:182, 1:26]
y_test = test.iloc[1:182, 26:27]
x_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
normal = StandardScaler()
x_new = normal.fit_transform(x_train)
x_train= pd.DataFrame(x_new, dtype='float')
x_new = normal.fit_transform(x_test)
x_test= pd.DataFrame(x_new, dtype='float')
x_test.head()
```

```python
from sklearn.svm import SVC  
svclassifier = SVC(kernel='linear')  
svclassifier.fit(x_train, y_train.values.ravel())
```

```python
y_pred = svclassifier.predict(x_test) 
```

```python
svclassifier.score(x_test,y_test)
```
